```python
import seaborn as sns
import pandas as pd  
import matplotlib.pyplot as plt  
sns.set_theme()
df = pd.read_csv("Football.csv")
```

```python
plt.hist(df['Goals'])
plt.xlabel('Number of Goals Scored')
plt.ylabel('Number of Players')
plt.title('Distribution of Goals Scored')
plt.show()
```

```python
sns.scatterplot(
    x="Age",
    y="Appearances",
    data=df
)

# Customize the plot using plt (optional)
plt.xlabel('Age')
plt.ylabel('Number of Appearances')
plt.title('Relationship between Age and Appearances')
plt.show()
```

```python
sns.boxplot(
    x = "Position",
    y = "Shooting accuracy %",  
    showmeans=True,  
    data=df
)
plt.xlabel('Position')
plt.ylabel('Shooting Accuracy (%)')
plt.title('Distribution of Shooting Accuracy by Position')
plt.show()
```

```python
sns.pairplot(df) 
plt.show()  
```
